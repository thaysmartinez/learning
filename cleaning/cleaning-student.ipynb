#### Three variables in two columns in `treatments` table (treatment, start dose and end dose)

##### Define
Transform columns auralin and novodra into values under treatments column. Create two new columns (variables) to store start and end doses. Remove string 'u' from data and convert to integers.

Hint: use pandas' [melt function](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.melt.html) and [`str.split()` method](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.str.split.html). Here is an excellent [`melt` tutorial](https://deparkes.co.uk/2016/10/28/reshape-pandas-data-with-melt/).*

##### Code

```python
# Your cleaning code here
treatments_clean = pd.melt(treatments_clean, id_vars=['given_name', 'surname', 'hba1c_start', 'hba1c_end', 'hba1c_change'], var_name='treatment', value_vars=['auralin', 'novodra'])

# Drop rows which treatment values are '-'
index_names = treatments_clean[treatments_clean['value'] == '-'].index
treatments_clean.drop(index_names, inplace=True)

treatments_clean.reset_index(drop=True,inplace=True)

# Split value column by '-' delimiter
treatments_clean[['start_dose', 'end_dose']] = treatments_clean.value.str.split('-', expand=True)

# Remove 'value' column
treatments_clean.drop(columns=['value'], inplace=True)

# Remove character 'u' from start_dose and end_dose columns and convert them to integer
treatments_clean['start_dose'] = treatments_clean['start_dose'].str.replace('u', '').astype(int)
treatments_clean['end_dose'] = treatments_clean['end_dose'].str.replace('u', '').astype(int)

# Reorder columns to original structure
ordered_columns = (['given_name', 'surname', 'treatment', 'start_dose', 'end_dose', 
                    'hba1c_start', 'hba1c_end', 'hba1c_change'])

treatments_clean = treatments_clean[ordered_columns]
```

##### Define
*Your definition here. Hint: [tutorial](https://chrisalbon.com/python/pandas_join_merge_dataframe.html) for one function used in the solution and [tutorial](http://erikrood.com/Python_References/dropping_rows_cols_pandas.html) for another function used in the solution.*

##### Code

```python
# Capitalize first character on given_name and surname columns
treatments_clean['given_name'] = treatments_clean['given_name'].str.title()
treatments_clean['surname'] = treatments_clean['surname'].str.title() 

# Merge patient_id column from patients_clean into treatments_clean
treatments_clean = pd.merge(treatments_clean, patients_clean[['patient_id', 'given_name', 'surname']], on=('given_name', 'surname'), how='left')

# Remove given_name and surname columns
treatments_clean.drop(['given_name', 'surname'], axis=1, inplace=True)

# Reorder columns in treatments_clean DataFrame
ordered_columns = (['patient_id', 'treatment', 'start_dose', 'end_dose', 'hba1c_start',
                    'hba1c_end', 'hba1c_change', 'adverse_reaction'])

treatments_clean = treatments_clean[ordered_columns]
```

```python
# Capitalize first character on given_name and surname columns
adverse_reactions_clean['given_name'] = adverse_reactions_clean['given_name'].str.title()
adverse_reactions_clean['surname'] = adverse_reactions_clean['surname'].str.title()

# Merge patient_id column from patients_clean into adverse_reactions_clean
adverse_reactions_clean = pd.merge(adverse_reactions_clean, 
                                   patients_clean[['patient_id', 'given_name', 'surname']], 
                                   on=('given_name', 'surname'), how='left')

# Remove given_name and surname_columnss
adverse_reactions_clean.drop(['given_name', 'surname'], axis=1, inplace=True)

# Reorder columns in adverse_reactions_clean DataFrame
adverse_reactions_clean = adverse_reactions_clean[['patient_id', 'adverse_reaction']]
```
